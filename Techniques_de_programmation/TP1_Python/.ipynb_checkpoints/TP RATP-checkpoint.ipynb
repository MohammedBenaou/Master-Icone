### Question 4

Utiliser une compréhension de liste pour n'extraire que les lignes qui contiennent la chaine `<dataset:`. Mettre le résultat dans la variable `dataset`.

### Question 5

Créer un dictionnaire `station` contenant :
- le nom de la station
- le réseau concerné
- le trafic de la station
- la ville dans laquelle se trouve la station

> **Challenge** : en se débrouillant bien tout peut être fait en une ligne avec une compréhension de dictionnaire.

```python
def func_return_station(data):
    stat={}
    for ds in data:  
        if "station" in ds:
            stat["nom"]=ds[ds.find(">")+1:ds.find("</")]
        if "reseau" in ds:
            stat["reseau"]=ds[ds.find(">")+1:ds.find("</")]
        if "trafic" in ds:
            stat["trafic"]=ds[ds.find(">")+1:ds.find("</")]
        if "ville" in ds:
            stat["ville"]=ds[ds.find(">")+1:ds.find("</")]
    return stat

station={}
station=func_return_station(dataset)
print (station) 
```

```python
with open ('donnees10.rss') as c:
    contenu=c.read()
data_contenue=contenu.split('\n')
list_of_data={}
co=0
enditems=[]
for _li in data_contenue:    
    if "</item" in _li:
        enditems.append(co)
    co=co+1    
cpt=0
list_of_data[cpt]=func_return_station(data_contenue[9:35])
i=1
while i<len(enditems)-1:    
    list_of_data[i]=func_return_station(data_contenue[enditems[i]:enditems[i+1]])
    i=i+1
print(list_of_data)   
```

### Question 7

Afin de rendre le traitement plus générique. Nous allons créer un fonction **récursive** permettant de construire la liste de dictionnaires des stations. 

Cette fonction aura pour prototype `def parse(contenu,liste)` : avec `contenu` la chaine contenant le reste de la chaine à parser (au début toute la chaine - le fichier ayant été lu préalablement), et `liste` étant la liste qui doit contenir le résultat (préalablement initialisé).

```python
def parse(contenu, liste):
    data = contenu[contenu.find("<item>")+6:]
    data = data[:-(len(contenu) - contenu.find("</item>"))]
    contenu = contenu[contenu.find("</item>")+6:]
    
    if(contenu.find("<item>")!=-1):
        liste = parse(contenu, liste)
        
    lines = data.splitlines()
    for i, _ in enumerate(lines):
        lines[i] = lines[i].strip()
    
    dataset = []
    for s in lines:
        if "<dataset:" in s:
            dataset.append(s)
            
    station = {}
    for l in dataset:
        if "station" in l:
            station["nom"] = l[17:-18]
        if "reseau" in l:
            station["reseau"] = l[16:-17]
        if "trafic" in l:
            station["trafic"] = l[16:-17]
        if "ville" in l:
            station["ville"] = l[15:-16]
            
    liste.append(station)
    
    return liste;

liste = []
contenu = open('donnees10.rss', "r")
contenu = contenu.read()
parse(contenu, liste)
    
```

### Question 10

Plutôt que de télécharger les informations au fur et à mesure, comme on l'a fait jusqu'à présent, il serait bien plus facile d'aller les chercher directement en ligne quand on en a besoin. D'autant plus quand les informations sont disponibles en temps réel. C'est ce que l'on va faire en utilisant le package [`urllib`](https://docs.python.org/3/library/urllib.html).

Outre les informations sur 2016, la RATP dispose aussi des archives pour 2013, 2014 et 2015. Il suffit encore une fois seulement de modifier l'URL : https://data.ratp.fr/api/records/1.0/search/?dataset=trafic-annuel-entrant-par-station-du-reseau-ferre-"+ANNEE+&rows="+NBLIGN

> **Attention** : pour l'année 2013 l'URL est différente : https://data.ratp.fr/api/records/1.0/search/?dataset=trafic-annuel-entrant-par-station-du-reseau-ferre"+&rows="+NBLIGN

Utiliser `urllib` et `json` pour construire un dictionnaire contenant les `fields` de tous les `records` sur toutes les années (les clefs seront les années).

```python
import urllib
import json
listeannee=[2013,2014,2015,2016]
data={}

for anne in listeannee:
    if anne==2013:
        url="https://data.ratp.fr/api/records/1.0/search/?dataset=trafic-annuel-entrant-par-station-du-reseau-ferre&rows=10000"
    else:
        url = "https://data.ratp.fr/api/records/1.0/search/?dataset=trafic-annuel-entrant-par-station-du-reseau-ferre-"+str(anne)+"&rows=10000"   
    respons = urllib.request.urlopen(url)
    inter = json.loads(respons.read().decode(respons.info().get_param('charset') or 'utf-8'))
    inter = inter["records"]
    data[anne] = [d['fields'] for d in inter]
    
data
```
